```python
import pandas as pd
url = "https://raw.githubusercontent.com/Estadistica2025BB/Estadistica2025B/refs/heads/main/Furniture.csv"
df = pd.read_csv(url)
df = df.dropna()
df
```

```python
df.columns
```

```python
# En este primer bloque cargamos la base de datos directamente desde GitHub.
# La información contiene variables relacionadas con precios, costos, inventarios,
# descuentos y características de distintos muebles.
# Luego se eliminan posibles valores nulos y se renombran las columnas al español
# para facilitar la interpretación.

df.columns = ['precio', 'costo', 'ventas',
       'margen_beneficio', 'inventario', 'porcentaje_descuento',
       'días_entrega', 'categoria', 'material', 'color',
       'ubicación', 'estación', 'tipo_tienda', 'marca', 'ganancia']
df
```

```python
# Este bloque muestra gráficos de pastel (pie charts) para las variables categóricas.
# Permite observar la proporción de productos por tipo, material, color, marca, etc.
# Así se identifican las categorías más comunes en el conjunto de datos.

import matplotlib.pyplot as plt

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print(categorical_cols)

for col in categorical_cols:
    fig, ax = plt.subplots(figsize=(5, 5))

    df[col].value_counts(dropna=False).plot.pie(
        autopct='%1.1f%%',
        ax=ax,
        shadow=False,
        textprops={'fontsize': 12}
    )
    ax.set_title(f'Gráfico de pastel para la variable "{col}"', fontsize=12)
    ax.set_ylabel('')

    plt.tight_layout()
    plt.show()
```

**D) — Influencia de la categoría del producto:**

La categoría tiene una influencia significativa en los precios y volúmenes de venta. Los resultados del ANOVA muestran diferencias notables (p < 0.001), aunque la prueba de Tukey indica que no todas las categorías difieren significativamente entre sí. Esto sugiere que, aunque las categorías afectan el precio promedio, las variaciones dentro de cada grupo no son extremas.

```python
# En este bloque se aplica un modelo lineal general y una tabla ANOVA (análisis de varianza)
# para evaluar qué variables explican mejor el precio.
# ANOVA permite determinar si existen diferencias estadísticamente significativas
# entre los promedios de distintas categorías o factores.

from statsmodels.formula.api import ols
import statsmodels.api as sm
# Y ~ X
modelo_lineal = ols(' precio ~ costo + ventas + margen_beneficio + inventario + porcentaje_descuento + días_entrega + ganancia +  C(categoria) + C(material) + C(color) + C(ubicación) + C(estación) + C(tipo_tienda) + C(marca) + (ganancia)',
                    data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

**C) — Relación entre costo, inventario y precio de venta:**

Existe una relación fuertemente positiva entre el costo y el precio (p < 0.001), indicando que los productos más caros de fabricar tienden a venderse a precios mayores.
Por otro lado, el inventario no muestra relación significativa (p ≈ 0.59), lo cual sugiere que la cantidad disponible no influye directamente en el precio de venta.

```python
# La prueba de Tukey sirve para comparar los promedios entre grupos
# y determinar cuáles categorías de productos difieren significativamente en precio.

from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['precio'], groups=df['tipo_tienda'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

```python
df["categoria"]= df["categoria"].map({"Bed":0, "Chair":1, "Desk": 2, "Sofa": 3, "Table": 4})
df

df["material"]= df["material"].map({"Glass":0, "Plastic":1, "Metal": 2, "Wood": 3, "Fabric": 4})
df

df["color"]= df["color"].map({"Green": 0, "Blue": 2, "Brown": 3, "Red":4,  "Black": 5 ,"White": 6})
df

df["ubicación"]= df["ubicación"].map({"Urban": 0, "Rural":1, "Suburban":2})
df

df["estación"]= df["estación"].map({"Winter": 0, "Summer":1, "Spring":2, "Fall": 3})
df


df["marca"]= df["marca"].map({"BrandA":0, "BrandB":1, "BrandD": 2, "BrandC": 3})
df
```

```python
# En este paso se calcula la matriz de correlación entre todas las variables numéricas.
# Se utiliza un mapa de calor (heatmap) para visualizar la fuerza y dirección
# de las relaciones entre variables.
# Los valores cercanos a 1 indican relación positiva; los cercanos a -1, negativa

import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", square=True, center=0, cbar_kws={"shrink": .8})
plt.title('Mapa de Calor de la Matriz de Correlación')
plt.show()
```

A) — Relación entre precio de venta y volumen de ventas:
La correlación entre precio y ventas es negativa. Esto significa que, en general, a medida que el precio aumenta, el volumen vendido tiende a disminuir.
Este comportamiento refleja la lógica de la demanda: los consumidores compran menos cuando el precio sube.


B) — Relación entre porcentaje de descuento y volumen de ventas:
El porcentaje de descuento no muestra una relación fuerte con las ventas. Esto indica que los descuentos aplicados no generaron un incremento sustancial en el volumen vendido

```python
# En este bloque se construye un modelo de regresión lineal múltiple (OLS).
# El objetivo es predecir el precio a partir de distintas variables explicativas.
# OLS (Ordinary Least Squares) estima una ecuación lineal que minimiza los errores al cuadrado.


x = df[[ 'costo', 'ventas',
       'margen_beneficio', 'inventario', 'porcentaje_descuento',
       'días_entrega', 'categoria', 'material', 'color',
       'ubicación', 'estación', 'marca', 'ganancia']]
y = df["precio"]

# 6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de
# dispersión.
import statsmodels.api as sm
x_constante = sm.add_constant(x)
modelo = sm.OLS(y, x_constante).fit()

yc = modelo.predict(x_constante)



# plt.plot(x, yc, color = 'black', linestyle = '--')

# 5. Calcule el coeficiente de determinación e interprete el resultado.
from sklearn.metrics import r2_score  # recomendada
r2 = r2_score(y, yc)
print(f'Coeficiente de determinación: {r2: 0.4f}\n')




# Aquí se analizan los residuales del modelo (diferencias entre valores observados y estimados).
# También se aplican pruebas estadísticas de normalidad (Shapiro) y homocedasticidad (Breusch-Pagan)
# para comprobar si los supuestos del modelo lineal se cumplen.


# 8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente,
# ¿Parece que se verifican los supuestos?

residuales = modelo.resid
plt.figure()
plt.scatter(x["costo"], residuales, color = '#008B8B')
plt.xlabel('costo', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)

residuales = modelo.resid
plt.figure()
plt.scatter(x["ventas"], residuales, color = '#008B8B')
plt.xlabel('ventas', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)

residuales = modelo.resid
plt.figure()
plt.scatter(x["margen_beneficio"], residuales, color = '#008B8B')
plt.xlabel('margen_beneficio', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)

residuales = modelo.resid
plt.figure()
plt.scatter(x["inventario"], residuales, color = '#008B8B')
plt.xlabel('inventario', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)

residuales = modelo.resid
plt.figure()
plt.scatter(x["porcentaje_descuento"], residuales, color = '#008B8B')
plt.xlabel('porcentaje_descuento', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)


residuales = modelo.resid
plt.figure()
plt.scatter(x["días_entrega"], residuales, color = '#008B8B')
plt.xlabel('días_entrega', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)

residuales = modelo.resid
plt.figure()
plt.scatter(x["ganancia"], residuales, color = '#008B8B')
plt.xlabel('ganancia', fontsize = 12)
plt.ylabel('precio', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales de Muebles", fontsize = 16)


# 9. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f'valor-p de Shapiro: {valor_p_sh: 0.4f}\n')

# 10. Realice la prueba de Brausch-Pagan para los residuales y comente el
# resultado.
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')

# 12. Realice una tabla ANOVA e interprete el resultado.
from statsmodels.formula.api import ols
# Y ~ X
modelo_lineal = ols('precio ~ costo + ventas + margen_beneficio + inventario + porcentaje_descuento + días_entrega + C(categoria) + C(material) + C(color) + C(ubicación) + C(estación) + C(marca) + (ganancia)', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```
